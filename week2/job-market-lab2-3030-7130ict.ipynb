```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load job market
df = pd.read_csv('job-market(3).csv')

df.head(5)
```

```python
jobsdict = dict()

df = df.dropna(subset=["Classification"])

for i in df["Classification"]:
    try:
        jobsdict[i] += 1
    except:
        jobsdict[i] = 1
    
jobSer = pd.Series(data=jobsdict)        
jobs = pd.DataFrame(jobSer, columns="Count".split())

jobs = jobs.sort_values('Count')

jobs
```

```python
%matplotlib inline

sns.barplot(jobs, x=jobs.iloc[:,0], y=jobs.index)
```

```python
import matplotlib
#create dataframe for salary ranges
df['SalaryRange'] = df['HighestSalary'] - df['LowestSalary']

df.head(10)
```

```python
saldict = dict()

for i in df["SalaryRange"]:
    try:
        saldict[i] += 1
    except:
        saldict[i] = 1
    
salSer = pd.Series(data=saldict)        
salaries = pd.DataFrame(salSer, columns="Count".split())

salaries = salaries.sort_values('Count')
#salary ranges plot

fig, ax = plt.subplots()
ax.pie(salaries.iloc[:,0], labels=salaries.index)
ax.set(title = "Job Posts by Salary Range")


plt.show()
```

```python
saldict = dict()

for i in df["LowestSalary"]:
    try:
        saldict[i] += 1
    except:
        saldict[i] = 1
    
salSer = pd.Series(data=saldict)        
salaries = pd.DataFrame(salSer, columns="Count".split())

salaries = salaries.sort_values('Count')
#salary ranges plot

fig, ax = plt.subplots()
ax.pie(salaries.iloc[:,0], labels=salaries.index)
ax.set(title = "Job Posts by Lowest Salary")


plt.show()
```
